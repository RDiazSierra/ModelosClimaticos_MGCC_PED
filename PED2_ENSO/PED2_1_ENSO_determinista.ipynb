## ENSO como problema de dinámica interna

El fenómeno El Niño–Oscilación del Sur (ENSO) se manifiesta como una variabilidad organizada en la temperatura superficial del océano, los vientos alisios y la convección tropical en el Pacífico ecuatorial, con escalas temporales interanuales.

Un rasgo fundamental de ENSO es que no responde a un forzamiento externo periódico que imponga su ritmo. Esto sugiere que su carácter oscilatorio surge, en gran medida, de la **dinámica interna del sistema océano–atmósfera**, es decir, de la interacción entre sus propios componentes.

Desde un punto de vista físico, esto plantea una pregunta central:  
*¿cómo puede un sistema climático generar oscilaciones coherentes a partir de la interacción de sus propios componentes, sin necesidad de una señal externa periódica?*

La figura siguiente resume, de forma esquemática, las configuraciones características del Pacífico ecuatorial asociadas a las fases principales de ENSO:

<img src="figuras/EsquemaENSO.png" alt="Esquema conceptual del ENSO">

En condiciones **neutras**, los vientos alisios soplan de este a oeste, empujando el agua superficial cálida hacia el Pacífico occidental. Como consecuencia, la termoclina se inclina: es profunda en el oeste y somera en el este, donde el afloramiento de aguas frías mantiene temperaturas superficiales relativamente bajas. Esta configuración permite al océano almacenar calor de forma desigual a lo largo de la cuenca.

Durante episodios de **El Niño**, los vientos alisios se debilitan y la circulación atmosférica se reorganiza. El agua cálida superficial se extiende hacia el Pacífico central y oriental, la termoclina se aplana y el afloramiento en el este se reduce. Esto favorece un calentamiento de la superficie oceánica y una liberación más eficiente de calor hacia la atmósfera, con consecuencias climáticas a escala global.

En fases de **La Niña**, ocurre el proceso opuesto. Los alisios se intensifican, el transporte de agua cálida hacia el oeste aumenta y la termoclina se inclina aún más. El afloramiento de aguas frías en el este se refuerza, enfriando la superficie, mientras que el océano acumula calor en capas subsuperficiales, especialmente en el Pacífico occidental.

Estos esquemas ponen de manifiesto que ENSO no es simplemente una alternancia de estados superficiales, sino el resultado de un **intercambio continuo de energía entre el océano y la atmósfera**, mediado por procesos que actúan en escalas temporales distintas. Comprender cómo esta interacción puede dar lugar a oscilaciones internas es el objetivo central del enfoque dinámico que se desarrolla a continuación.



## Condiciones físicas del Pacífico ecuatorial

ENSO no se manifiesta en cualquier cuencas oceánicas. Su existencia está estrechamente ligada a las características específicas del Pacífico ecuatorial, entre ellas:

- su gran extensión zonal (de este a oeste),
- la circulación ecuatorial forzada por los vientos alisios,
- la asimetría este–oeste en la temperatura superficial,
- la presencia de afloramiento frío en el Pacífico oriental,
- y una termoclina inclinada, profunda en el oeste y somera en el este.

Estas características permiten que el océano almacene y redistribuya calor de forma desigual (en profundidad  y en diferentes zonas del eje E-OE) y que los ajustes dinámicos no se produzcan todos en la misma escala temporal. En particular, la interacción entre procesos rápidos (superficie y atmósfera) y procesos más lentos (interior del océano) introduce desfases temporales que resultan esenciales para la aparición de variabilidad organizada.


## De la estructura espacial a los modos dinámicos

Aunque ENSO suele describirse mediante patrones espaciales este–oeste (temperaturas, vientos, afloramiento), su dinámica puede entenderse también desde una perspectiva más agregada.

En esta aproximación, no se describe explícitamente la geometría de la cuenca ni las corrientes detalladas, sino que se identifican **componentes dinámicos del sistema** que responden a esas escalas temporales distintas identificadas al caracterizar el fenómeno y que están acoplados entre sí.

Esta forma de pensar es habitual en modelización: se parte de descripciones más complejas y se realizan simplificaciones y promedios que conservan únicamente los grados de libertad necesarios para reproducir el comportamiento dinámico esencial.


## Modelos conceptuales y comprensión dinámica

Los modelos climáticos completos reproducen ENSO como un fenómeno emergente de múltiples procesos acoplados. Sin embargo, su complejidad dificulta identificar y entender con claridad qué mecanismos son esenciales para la aparición de la oscilación.

Los modelos conceptuales mecanicistas no buscan reproducir todos los detalles del fenómeno, sino aislar los procesos dominantes y analizar sus consecuencias dinámicas. Su valor principal es explicativo: permiten explorar cómo determinadas estructuras de acoplamiento pueden dar lugar a oscilaciones internas, incluso en ausencia de forzamientos externos periódicos.

En este contexto, resulta útil trabajar con modelos muy simplificados del ENSO que representen el “esqueleto dinámico” de este sistema.


## Variables del modelo: interpretación física

El modelo del oscilador de recarga describe la dinámica del sistema océano–atmósfera tropical mediante dos variables agregadas:

- $T(t)$ (tonos rojizos en los diagramas) representa la anomalía de la temperatura superficial del océano en el Pacífico ecuatorial, directamente relacionada con las manifestaciones observables de ENSO.
- $h(t)$ (tonos amarillentos en los diagramas) representa un modo lento asociado al contenido de calor oceánico efectivo, estrechamente relacionado con la profundidad media de la termoclina y el almacenamiento de calor en capas subsuperficiales.

Estas variables **no describen compartimentos físicos independientes ni regiones espaciales cerradas**. Ambas coexisten en el sistema real y deben interpretarse como **modos dinámicos con escalas temporales distintas**, cuya existencia está permitida por la geometría y la circulación características del Pacífico ecuatorial.

La figura siguiente ilustra de forma cualitativa cómo estas dos variables pueden interpretarse en las distintas fases de ENSO:


<img src="figuras/EsquemaENSO_variables.png" width="900" alt="Esquema conceptual de las variables del modelo ENSO">

En este esquema, las zonas coloreadas **no representan áreas ni volúmenes físicos**, sino que indican de manera cualitativa el signo y la intensidad relativa de las anomalías asociadas a $T(t)$ y $h(t)$ en cada fase. Durante episodios tipo El Niño, las anomalías de temperatura superficial dominan, mientras que el contenido de calor subsuperficial tiende a descargarse. En fases tipo La Niña, ocurre lo contrario: el océano acumula calor en profundidad y las anomalías superficiales son más débiles o de signo opuesto.

El estado neutro corresponde a una situación intermedia, en la que ninguno de los dos modos domina claramente. Este tipo de representación permite conectar la descripción física del fenómeno con las variables dinámicas utilizadas en el modelo, sin introducir una interpretación espacial literal.



## Estructura dinámica del modelo e interpretación de las interacciones

La evolución conjunta de $T(t)$ y $h(t)$ se describe mediante un sistema de ecuaciones diferenciales ordinarias lineales:

$$
\frac{dT}{dt} = -a_{11}\,T + a_{12}\,h
$$

$$
\frac{dh}{dt} = -a_{22}\,h - a_{21}\,T
$$

En este sistema, los coeficientes $a_{ij}$ representan, de forma agregada, amortiguamientos y acoplamientos efectivos entre los dos modos dinámicos del sistema. No corresponden açprocesos físicos individuales, sino al efecto combinado de múltiples mecanismos océano–atmósfera que han sido promediados y simplificados.

La figura siguiente resume gráficamente la estructura de interacciones del modelo:

<img src="figuras/ENSOInteracciones.png" width="400" alt="Esquema conceptual de las interacciones del modelo ENSO">

Los términos $-a_{11}T$ y $-a_{22}h$ representan la tendencia de cada variable a disipar anomalías en ausencia de acoplamiento. En términos físicos, recogen de forma muy simplificada la pérdida de anomalías de temperatura superficial por intercambio con la atmósfera y la relajación del contenido de calor oceánico hacia un estado medio, respectivamente.

Los términos cruzados describen el acoplamiento entre ambos modos dinámicos. El término proporcional a $h$ en la ecuación de $T$ indica que variaciones en el contenido de calor oceánico influyen en la temperatura superficial, mientras que el término proporcional a $T$ en la ecuación de $h$ recoge el efecto inverso: cambios en la temperatura superficial modifican, de forma agregada, la evolución del contenido de calor oceánico.

La interacción entre estos dos procesos genera una retroalimentación dinámica. Dado que $T(t)$ y $h(t)$ evolucionan en escalas temporales distintas, el sistema no responde de manera instantánea a las perturbaciones, sino que introduce desfases temporales en su evolución. Bajo determinadas condiciones en los valores de los parámetros, esta estructura puede dar lugar a comportamientos oscilatorios, interpretables como una manifestación de variabilidad interna del sistema.



## Escalas temporales y alcance del enfoque

Un rasgo clave del modelo es la diferencia entre las escalas temporales asociadas a $T(t)$ y $h(t)$. Esta diferencia permite que el sistema no responda de forma instantánea a las perturbaciones, sino que tienda a sobrecorregirse, favoreciendo la aparición de oscilaciones internas.

En última instancia, la ENSO existe porque el Pacífico ecuatorial permite ajustes rápidos que afectan principalmente a la superficie del océano y ajustes más lentos que redistribuyen el contenido de calor en el interior de la cuenca, y ambos están acoplados mediante un desfase temporal inevitable.
De un modo más técnico (y a un nivel algo superior al de este ejercicio) podemos decir que: 
- Los ajustes rápidos que afectan a la superficie están dominados por la interacción entre la circulación atmosférica tropical y la respuesta superficial del océano, y se manifiestan a través de cambios en los vientos alisios, en la temperatura superficial y en la profundidad de la termoclina, ajustados dinámicamente mediante ondas ecuatoriales de tipo Kelvin.
- Los ajustes lentos que redistribuyen el calor en el interior del océano están controlados por la respuesta de gran escala de la cuenca a esas mismas perturbaciones, y se producen mediante procesos dinámicos más lentos, asociados a ondas de Rossby ecuatoriales, que reorganizan progresivamente el contenido de calor subsuperficial y la estructura media de la termoclina.

Es importante subrayar que el modelo no describe explícitamente procesos espaciales como corrientes este–oeste, afloramiento o propagación de ondas ecuatoriales. Estos procesos están implícitos en la estructura del modelo y en los valores efectivos de los parámetros.

Recordamos, de nuevo, que el objetivo no es reproducir ENSO de forma realista, sino comprender cómo una estructura dinámica mínima puede capturar aspectos esenciales de su comportamiento.


## Diagrama conceptual de interacciones

El diagrama se interpreta siguiendo un recorrido progresivo que va desde una descripción espacial cualitativa del fenómeno hasta su formulación como sistema dinámico de baja dimensión.

En la parte inferior izquierda se muestra un esquema espacial idealizado del Pacífico ecuatorial, que resume de forma cualitativa la circulación océano–atmósfera asociada a ENSO. En este esquema aparecen las anomalías de temperatura superficial del océano, la estructura de la termoclina y los intercambios entre atmósfera y océano. Su función es únicamente proporcionar una referencia física y espacial del fenómeno; estos patrones no se representan explícitamente en el modelo.

A partir de esta descripción espacial se introducen, en la parte superior izquierda, las variables dinámicas del modelo. La variable $T(t)$, representada en tonos rojizos, corresponde a una medida agregada de las anomalías de temperatura superficial del océano en la región ecuatorial. La variable $h(t)$, representada en tonos anaranjados, resume el contenido de calor oceánico asociado a la estructura de la termoclina. Ambas variables condensan procesos espaciales complejos en magnitudes escalares dependientes del tiempo.

La parte derecha del diagrama muestra cómo estas variables se acoplan dinámicamente. La anomalía de temperatura superficial $T(t)$ influye sobre la evolución del contenido de calor oceánico $h(t)$ a través de un término de acoplamiento, y a su vez $h(t)$ retroalimenta la evolución de $T(t)$. Estos acoplamientos, representados mediante flechas orientadas, constituyen una estructura de realimentación que es central para la dinámica del sistema. Al mismo tiempo, cada variable está sujeta a un término de amortiguamiento que limita su crecimiento y representa la disipación efectiva de las anomalías.

Finalmente, en la parte inferior derecha se enfatiza el papel de las escalas temporales. La evolución de $T(t)$ está asociada a procesos relativamente rápidos, mientras que $h(t)$ responde en escalas más largas. Esta diferencia en las tasas de cambio introduce un desfase dinámico entre ambas variables, que resulta esencial para la aparición de oscilaciones organizadas en el sistema, incluso en ausencia de variabilidad en los forzamientos externos.


```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# Parámetros del modelo (tiempo en meses)
# Nota: estos valores están pensados para exploración conceptual, no para ajuste cuantitativo a datos.

params = {
    "a11": 0.076,   # amortiguamiento efectivo de T (mes^-1)
    "a12": 0.0236,  # acoplamiento h -> T (mes^-1, h en "unidades" del modelo)
    "a21": 1.25,    # acoplamiento T -> h (mes^-1)
    "a22": 0.008    # amortiguamiento efectivo de h (mes^-1)
}
```

```python
def recharge_oscillator_rhs(t, y, params):
    """
    Sistema determinista de baja dimensión tipo 'oscilador de recarga' para ENSO.

    Variables:
    - y[0] = T(t): anomalía de temperatura superficial (proxy ENSO)
    - y[1] = h(t): variable oceánica de evolución más lenta (proxy contenido de calor/termoclina)

    Tiempo:
    - t en meses
    """
    T, h = y
    a11 = params["a11"]
    a12 = params["a12"]
    a21 = params["a21"]
    a22 = params["a22"]

    dTdt = -a11 * T + a12 * h
    dhdt = -a21 * T - a22 * h

    return [dTdt, dhdt]
```

```python
# Horizonte temporal (meses)
years = 80
t0 = 0.0
t1 = 12.0 * years

# Puntos donde queremos la salida (para graficar de forma estable)
dt_out = 0.25  # meses (salida cada ~7.5 días)
t_eval = np.arange(t0, t1 + dt_out, dt_out)

# Condiciones iniciales (puedes ajustarlas si quieres explorar sensibilidad)
T0 = 0.5   # K (anomalía inicial)
h0 = 0.0   # unidades del modelo
y0 = [T0, h0]
```

```python
sol = solve_ivp(
    fun=lambda t, y: recharge_oscillator_rhs(t, y, params),
    t_span=(t0, t1),
    y0=y0,
    t_eval=t_eval,
    method="RK45",
    rtol=1e-8,
    atol=1e-10
)

if not sol.success:
    raise RuntimeError("La integración ha fallado: " + sol.message)

t = sol.t
T = sol.y[0]
h = sol.y[1]
```

```python
# Ventana final a mostrar (años)
years_to_show = 30
t_start_plot = t1 - 12.0 * years_to_show

mask = t >= t_start_plot
t_plot = t[mask] / 12.0  # convertir a años solo para la gráfica
T_plot = T[mask]
h_plot = h[mask]

plt.figure(figsize=(10, 4))
plt.plot(t_plot, T_plot, label="T(t) (anomalía SST)")
plt.plot(t_plot, h_plot, label="h(t) (variable oceánica lenta)")
plt.xlabel("Tiempo (años)")
plt.ylabel("Magnitud (unidades del modelo)")
plt.title("Simulación determinista del oscilador de recarga")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### Identificación cualitativa de las fases de ENSO

En la figura se muestran las evoluciones temporales de $T(t)$ (anomalía de temperatura superficial del océano) y $h(t)$ (variable oceánica lenta asociada al contenido de calor).

Aunque estas variables no corresponden directamente a observables individuales, $T(t)$ puede relacionarse cualitativamente con índices reales de SST (por ejemplo, Niño3.4), y $h(t)$ con indicadores integrados del contenido de calor oceánico, como el *Warm Water Volume* o anomalías de nivel del mar. Esta comparación debe entenderse únicamente en términos cualitativos.

### Interpretación dinámica del diagrama de fase

Este diagrama representa la trayectoria del sistema en el espacio de fases $(T,h)$, es decir, la evolución conjunta de la anomalía de temperatura superficial y de la variable oceánica lenta en tiempos muy largos.

En ausencia de forzamientos externos o perturbaciones adicionales, el sistema no presenta un ciclo límite autosostenido siempre: la trayectoria describe una espiral que converge progresivamente hacia el estado de equilibrio. Este comportamiento es análogo al *spin-up* estudiado en actividades anteriores, en las que el sistema ajusta sus variables internas hasta alcanzar un régimen final  determinado por su estructura dinámica.

La presencia de trayectorias espirales indica que el acoplamiento entre $T$ y $h$ introduce desfases temporales y una tendencia oscilatoria, pero que los términos de amortiguamiento dominan a largo plazo. Como resultado, las oscilaciones se atenúan con el tiempo si el sistema permanece aislado.

En el contexto climático real, este resultado sugiere que la persistencia de la variabilidad tipo ENSO no requiere un forzamiento periódico específico, pero sí la acción continua de perturbaciones —asociadas a procesos no representados explícitamente en el modelo— que desplacen al sistema fuera del equilibrio. Estas perturbaciones pueden interpretarse como fluctuaciones inducidas en $T$ o $h$ por la atmósfera, la circulación oceánica, las ondas ecuatoriales u otros procesos internos del sistema climático.

El papel del modelo no es identificar la naturaleza concreta de estas perturbaciones, sino mostrar cómo, una vez activado, el acoplamiento dinámico entre escalas rápidas y lentas puede organizar la respuesta del sistema en forma de oscilaciones coherentes.


## Exploración guiada de parámetros

En esta sección se explora de forma controlada cómo cambia el comportamiento dinámico del sistema al modificar **un único parámetro** del modelo.

El objetivo no es realizar un estudio exhaustivo del espacio de parámetros, sino analizar qué rasgos del comportamiento dinámico son robustos y cuáles se ven afectados por cambios moderados en los coeficientes del sistema.

En particular, se explorará el efecto del parámetro $a_{11}$, asociado al amortiguamiento efectivo de la variable $T(t)$.


```python
# Valores a explorar del parámetro a11 (amortiguamiento de T)
a11_values = [0.09, 0.076, 0.06]

results = {}

for a11 in a11_values:
    # Copiamos el diccionario de parámetros y modificamos solo a11
    params_test = params.copy()
    params_test["a11"] = a11

    sol_test = solve_ivp(
        fun=lambda t, y: recharge_oscillator_rhs(t, y, params_test),
        t_span=(t0, t1),
        y0=y0,
        t_eval=t_eval,
        method="RK45",
        rtol=1e-8,
        atol=1e-10
    )

    if not sol_test.success:
        raise RuntimeError(f"Integración fallida para a11 = {a11}")

    results[a11] = {
        "t": sol_test.t,
        "T": sol_test.y[0],
        "h": sol_test.y[1]
    }
```

```python
plt.figure(figsize=(10, 4))

for a11, data in results.items():
    mask = data["t"] >= t_start_plot
    t_plot = data["t"][mask] / 12.0  # años
    T_plot = data["T"][mask]

    plt.plot(t_plot, T_plot, label=f"a11 = {a11}")

plt.xlabel("Tiempo (años)")
plt.ylabel("T(t) (anomalía SST)")
plt.title("Efecto del parámetro a11 sobre la dinámica del sistema")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
